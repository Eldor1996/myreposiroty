```python
import matplotlib.pyplot as plt
import numpy as np


t = np.arange(0, 11, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='rate',
       title='US Dollar change in the last two month')
ax.grid()

fig.savefig("test.png")
plt.show()
```
